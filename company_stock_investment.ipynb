```python
#testing stationarity
from statsmodels.tsa.stattools import adfuller
```

```python
oil_result=adfuller(df['Oil Investments'])
```

```python
print("The values given as output by adfuller is : \n 'ADF test static','p-value','#lags used','number of observation used'\n\n",oil_result)
```

P-value is lesser than null hypothesis ,so it is stationary

```python
gold_result=adfuller(df['Gold Investments'])
```

```python
print("The values given as output by adfuller is : \n 'ADF test static','p-value','#lags used','number of observation used'\n\n",gold_result)
```

P-value is lesser than null hypothesis so it is stationary

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
import statsmodels.api as sm
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
# First 12 values are null so we will start from 13
fig = sm.graphics.tsa.plot_acf(df['Oil Investments'],lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df['Oil Investments'],lags=40,ax=ax2)
```

```python
#ARIMA MODEL IMPLEMENT
from statsmodels.tsa.arima_model import ARIMA
```

```python
oil=ARIMA(df['Oil Investments'],order=(1,1,1))
oil_fit=oil.fit()
```

```python
oil_fit.summary()
```

```python
df['oil_forecast']=oil_fit.predict(start=2,end=1980,dynamic=True)
df[['Oil Investments','oil_forecast']].plot()
```
